```python
def find_max_element(lst):
    """
    Function to find the maximum element in a list.
    :param lst: List[int] -> List of integers
    :return: int -> The maximum element in the list
    """
    m = lst[0]
    n = len(lst)
    for i in range(1 , n):
        if m < lst[i]:
            m = lst[i]
        
    return m
        

lst = [3, 5, 2, 9, 6]
find_max_element(lst)
```
