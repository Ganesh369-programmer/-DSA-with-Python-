Example:

Input: arr = [5, 4, 3, 2, 1]
Output: False
 
Input: arr = [1, 3, 2, 4, 5]
Output: False
 
Input: arr = [1, 2, 3, 4, 5]
Output: True

```python
def is_sorted(arr):
    
    for i in range(len(arr) - 1):
        
        if arr[i] < arr[i+1]:
            None
        elif arr[i] > arr[i+1]:
            return False
    else:
        return True
    
arr = [1, 2, 3, 4, 5]
is_sorted(arr)
```
