```python
def int_to_binary(n):
    """
    Function to convert an integer to its binary representation.
    
    Parameters:
    n (int): The integer to convert.
    
    Returns:
    str: The binary representation of the integer.
    """
    # Your code here
    r = 0
    q = n
    l = []
    while True:
        q = q // 2
        r = q % 2
        l.append(r)
        if q == 0:
            break

    result = "".join(str(j) for j in l)
    return result

int_to_binary(10)

```
