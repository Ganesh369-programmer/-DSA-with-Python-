Input: "Python"
Output: "nohtyP"

```python
def reverse_string(s):
    """
    Function to return the reversed version of the input string.
    
    Parameters:
    s (str): The input string to be reversed.
    
    Returns:
    str: The reversed string.
    """
    # Your code here
    l = []
    l1 = []

    for i in s:
        l.append(i)

    for j in l[::-1]:
        l1.append(j)
    
    result = ("".join(l1))
    return result


reverse_string("hello")

    
```
