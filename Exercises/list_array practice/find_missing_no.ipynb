## 1Ô∏è‚É£ What is this formula?

This is a **mathematical formula** to find the sum of numbers:

```
0 + 1 + 2 + 3 + ... + n
```

(or sometimes 1 to n ‚Äî both use the same formula)

### Formula:

```
Sum = n √ó (n + 1) √∑ 2
```

## 2Ô∏è‚É£ Why does this formula exist? (INTUITION)

Let‚Äôs take a small number:

### Example: n = 5

Numbers:

```
0, 1, 2, 3, 4, 5
```

Now pair them:

```
0 + 5 = 5
1 + 4 = 5
2 + 3 = 5
```

How many such pairs?

```
n + 1 = 6
```

Total sum:

```
(5 √ó 6) / 2 = 15
```

‚úî Same as:

```
0+1+2+3+4+5 = 15
```
