```python
def searchRange(nums, target):
    # Implement your solution here

    # binary search finds only one value so when we have to find multiple value then we have to use multiple binary search
    left , right = 0 , len(nums)-1
    start_index = -1
    while left <= right:
        mid = left + (right - left) // 2

        if nums[mid] == target:
            start_index = mid
            right = mid - 1

        elif nums[mid] > target: 
            right = mid - 1
        elif nums[mid] < target:
            left = mid + 1
        

    left = 0
    right = len(nums) - 1

    last_index = -1
    while left <= right:
        mid = left + (right - left) // 2

        if (nums[mid] == target):
            last_index = mid
            left = mid + 1
        elif nums[mid] > target:
            right = mid - 1
        elif nums[mid] < target:
            left = mid + 1
        

        
        
    return [start_index , last_index]

nums = [5, 7, 7, 8, 8, 10]
target = 8

searchRange(nums , target )
```
