```python
def is_prime(n):
    """
    Function to check if a number is prime.
    
    Parameters:
    n (int): The number to check.
    
    Returns:
    bool: True if n is prime, False otherwise.
    """
    # Your code here
    if n <= 1:
        return False
    else:
        # is_prime = True
        for i in range(2 , int(n**0.5) + 1):
            if n % i == 0:
                return False
        
        return True
```
