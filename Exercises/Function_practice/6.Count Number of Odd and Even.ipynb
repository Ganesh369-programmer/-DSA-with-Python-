Example:

Input: lst = [1, 2, 3, 4, 5]
Output: (2, 3)

There are 2 even numbers: 2, 4

There are 3 odd numbers: 1, 3, 5

```python
def count_even_odd(lst):
    # Your code goes here
    c =()
    e = 0
    o = 0
    for i in lst:
        if i%2 == 0:
            e += 1
        else:
            o += 1
    l = list(c)
    l.append(e)
    l.append(o)
    new_c = tuple(l)
    return new_c

count_even_odd([1, 2, 3, 4, 5])

```
